# 🎯 Principio de Incertidumbre Saussure-Heisenberg

## La imposibilidad de conocer simultáneamente el paradigma y el sintagma

**Analogía fundamental:**

| Mecánica Cuántica | Lingüística Saussureana |
|-------------------|-------------------------|
| Posición (x) | Eje sintagmático (combinación) |
| Momento (p) | Eje paradigmático (selección) |
| Δx · Δp ≥ ℏ/2 | ΔSintagma · ΔParadigma ≥ ℏ_sem/2 |

**Implicación:** No se puede conocer con precisión absoluta tanto la posición de un signo en la frase (sintagma) como su capacidad de sustitución (paradigma).

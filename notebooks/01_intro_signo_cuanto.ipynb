# 🌌 Introducción al Signo-Cuanto

## Fusión Saussure × Mecánica Cuántica

Este notebook introduce los conceptos fundamentales de la fusión entre la semiótica estructural de Ferdinand de Saussure y la mecánica cuántica.

| Saussure | Mecánica Cuántica | Fusión |
|----------|-------------------|--------|
| Significante + Significado | Estado cuántico | **Signo-Cuanto** |
| Valor diferencial | Amplitud de probabilidad | **Ser por no ser** |
| Lengua (*langue*) | Espacio de Hilbert | **Sistema de potenciales** |
| Habla (*parole*) | Colapso por medición | **Emergencia de realidad** |

## 1. Importar el paquete

```python
# Añadir el directorio padre al path
import sys
sys.path.insert(0, '..')

from saussure_quantum import SignoCuanto, Langue
from saussure_quantum.operators import operador_diferencia, principio_negatividad
from saussure_quantum.collapse import colapso_parole, MedidorParole, ContextoEnunciativo
import numpy as np
import matplotlib.pyplot as plt

print("✅ Paquetes importados correctamente")
print(f"Versión de numpy: {np.__version__}")
```

## 2. Crear nuestro primer Signo-Cuanto

Un signo-cuanto representa una palabra en superposición: coexisten múltiples significados potenciales.

```python
# Crear un signo-cuanto con tres posibles significados
signo = SignoCuanto(
    significantes=["🌞 sol", "🌙 luna", "⭐ estrella"],
    amplitudes=[1, 0.5, 0.3]
)

print("=== MI PRIMER SIGNO-CUANTO ===")
print(signo)
print(f"\nVector de amplitudes: {signo.amplitudes}")
print(f"Norma del vector: {np.linalg.norm(signo.amplitudes):.3f} (debe ser 1)")
```

## 3. Probabilidades y visualización

```python
# Calcular probabilidades
print("=== PROBABILIDADES ===")
for sig in signo.significantes:
    prob = signo.probabilidad(sig) * 100
    print(f"{sig}: {prob:.1f}%")

# Visualizar
fig, ax = plt.subplots(figsize=(8, 4))

probs = [signo.probabilidad(s) for s in signo.significantes]
ax.bar(signo.significantes, probs, color=['gold', 'silver', 'navy'])
ax.set_ylabel('Probabilidad')
ax.set_title('Probabilidad de colapso a cada significado')
ax.set_ylim(0, 1)

for i, p in enumerate(probs):
    ax.text(i, p + 0.02, f'{p*100:.1f}%', ha='center')

plt.tight_layout()
plt.show()
```

## 4. El acto de parole (colapso)

Cuando un hablante "usa" la palabra, la superposición colapsa a un solo significado. Este es el momento en que **emerge la realidad semiótica**.

```python
# Simular múltiples actos de habla
signo_original = SignoCuanto(["🍎 manzana", "🍐 pera", "🍊 naranja"], [1, 1, 1])
medidor = MedidorParole()

print("=== ACTOS DE PAROLE (MÚLTIPLES REALIDADES) ===\n")
print(f"Estado inicial: superposición equitativa\n")

for i in range(15):
    estado_actual = SignoCuanto(signo_original.significantes.copy(), signo_original.amplitudes.copy())
    resultado, _ = medidor.medir(estado_actual)
    print(f"Acto {i+1:2d}: el hablante dijo '{resultado}'")

print(f"\n📊 ESTADÍSTICAS:")
for palabra, count in medidor.estadisticas.items():
    print(f"  {palabra}: {count} veces ({count/15*100:.0f}%)")
```

## 5. El principio de negatividad (Saussure puro)

Saussure dijo: *"En la lengua solo hay diferencias sin términos positivos"*. Un fonema /p/ ES por NO SER /b/, /t/, /k/.

```python
# Crear fonemas puros
fonema_p = SignoCuanto(["/p/", "/b/", "/t/"], [1, 0, 0])
fonema_b = SignoCuanto(["/p/", "/b/", "/t/"], [0, 1, 0])
fonema_t = SignoCuanto(["/p/", "/b/", "/t/"], [0, 0, 1])

# Aplicar operador diferencia
diferencia_pura = operador_diferencia([fonema_p, fonema_b, fonema_t])

print("=== PRINCIPIO DE NEGATIVIDAD ===\n")
print("Fonemas puros (cada uno es 'algo'):")
print(fonema_p)

print("\nDespués de aplicar el operador diferencia (ser por NO SER):")
print(diferencia_pura)

# Análisis de negatividad
analisis = principio_negatividad(diferencia_pura)
print(f"\n📝 Análisis:")
print(f"  Significante principal: {analisis['significante_principal']}")
print(f"  Negatividad total: {analisis['negatividad_total']:.3f}")
```

## 6. Contexto y sesgo (intencionalidad del hablante)

No todos los actos de parole son iguales. El contexto (situación, intención, ruido) modifica las probabilidades.

```python
# Crear contexto con sesgo
signo_contexto = SignoCuanto(["🔥 calor", "❄️ frío", "💧 templado"], [1, 1, 1])

# Contexto 1: neutral
ctx_neutral = ContextoEnunciativo()

# Contexto 2: sesgo hacia "calor"
ctx_caliente = ContextoEnunciativo(intencionalidad=[0.8, 0.1, 0.1])

print("=== INFLUENCIA DEL CONTEXTO ===\n")

for ctx_name, ctx in [("Neutral", ctx_neutral), ("Sesgo a calor", ctx_caliente)]:
    resultados = []
    for _ in range(20):
        estado = SignoCuanto(signo_contexto.significantes.copy(), signo_contexto.amplitudes.copy())
        res, _, _ = colapso_parole(estado, ctx)
        resultados.append(res)
    
    print(f"\n📢 Contexto {ctx_name}:")
    for sig in signo_contexto.significantes:
        count = resultados.count(sig)
        print(f"  {sig}: {count}/20 ({count*5}%)")
```

## 7. Fases complejas (interferencia semántica)

En mecánica cuántica, las amplitudes pueden ser números complejos. Esto permite **interferencia**: dos significados pueden cancelarse o reforzarse.

```python
# Crear signo con fases
signo_complejo = SignoCuanto(
    significantes=["🐕 perro", "🐈 gato"],
    amplitudes=[1, 1j]
)

print("=== SIGNO CON FASES COMPLEJAS ===")
print(signo_complejo)
print(f"\nFase relativa: {signo_complejo.fase_relativa(0, 1) * 180 / np.pi:.0f}°")

# Visualizar en el plano complejo
fig, ax = plt.subplots(figsize=(6, 6))
for i, sig in enumerate(signo_complejo.significantes):
    amp = signo_complejo.amplitudes[i]
    ax.arrow(0, 0, amp.real, amp.imag, head_width=0.05, head_length=0.05, 
             fc=['blue', 'red'][i], ec=['blue', 'red'][i], label=sig)
    ax.plot(amp.real, amp.imag, 'o', color=['blue', 'red'][i], markersize=10)

ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax.axvline(x=0, color='k', linestyle='-', alpha=0.3)
ax.set_xlabel('Parte Real')
ax.set_ylabel('Parte Imaginaria')
ax.set_title('Amplitudes en el plano complejo')
ax.legend()
ax.set_aspect('equal')
plt.show()
```

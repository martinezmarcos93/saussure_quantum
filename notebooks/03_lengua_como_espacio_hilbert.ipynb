# 🧠 La Lengua como Espacio de Hilbert

## El sistema de todas las diferencias potenciales

**Concepto clave:** La *langue* (lengua) de Saussure es análoga al espacio de Hilbert de la mecánica cuántica: un espacio de estados potenciales que solo se actualizan en el acto de habla.

En este notebook exploramos:
- El espacio vectorial de los significados posibles
- El producto interno como medida de similitud semántica
- Bases y dimensiones del sistema lingüístico
- Transformaciones entre representaciones

def cambiar_base(estado, matriz_transformacion):
    """Cambia la base de representación de un estado"""
    return matriz_transformacion @ estado.amplitudes

## 7. Evolución temporal: ¿cómo cambia la lengua?

Una ecuación análoga a Schrödinger para la evolución semántica.
"""

**CELDA 15 (Código):**
```python
def evolucion_semantica(estado_inicial, pasos=10, tasa_cambio=0.1):
    """Simula evolución semántica (análogo a ecuación de Schrödinger)"""
    estados = [estado_inicial.amplitudes.copy()]
    
    # Hamiltonianosimple (matriz de evolución con fases)
    d = estado_inicial.dimension
    H = np.diag(np.random.randn(d))  # Energías aleatorias
    
    for t in range(1, pasos):
        # Evolución unitaria U = exp(-iHt)
        U = np.linalg.expm(-1j * tasa_cambio * H * t)
        nuevo_estado = U @ estados[0]
        estados.append(nuevo_estado)
    
    return estados

# Crear estado inicial
lang_evol = Langue(3, "Lengua evolutiva")
lang_evol._base = ["pasado", "presente", "futuro"]
inicial = lang_evol.superposicion([1, 0, 0])  # Solo "pasado"

print("=== EVOLUCIÓN SEMÁNTICA DE LA LENGUA ===\n")
print("El significado de las palabras cambia con el tiempo\n")

# Evolucionar
evolucion = evolucion_semantica(inicial, pasos=20, tasa_cambio=0.05)

# Visualizar evolución
fig, ax = plt.subplots(figsize=(10, 6))

for i, tiempo in enumerate([0, 5, 10, 15, 19]):
    probs = np.abs(evolucion[tiempo])**2
    ax.plot(range(3), probs, marker='o', label=f't={tiempo}')

ax.set_xticks(range(3))
ax.set_xticklabels(["pasado", "presente", "futuro"])
ax.set_ylabel('Probabilidad')
ax.set_title('Evolución del significado en el tiempo')
ax.legend()
ax.grid(True, alpha=0.3)

plt.show()

print("\n📖 Interpretación:")
print("  El significado NO es fijo.")
print("  La lengua es un sistema dinámico.")
print("  Las palabras 'viajan' por el espacio de Hilbert lingüístico.")

## 7. Construyendo un Oráculo Semántico Cuántico

Vamos a construir un pequeño oráculo que "predice" el futuro colapsando superposiciones de eventos.
"""

**CELDA 15 (Código):**
```python
class OraculoCuantico:
    """Un oráculo que responde preguntas colapsando superposiciones"""
    
    def __init__(self):
        self.historial = []
    
    def preguntar(self, pregunta, opciones, sesgo=None):
        """Haz una pregunta al oráculo"""
        if sesgo is None:
            sesgo = [1] * len(opciones)
        
        # Crear superposición de respuestas posibles
        oraculo = SignoCuanto(opciones, sesgo)
        
        print(f"❓ Pregunta: {pregunta}")
        print(f"   Posibilidades: {opciones}")
        
        # Contexto con algo de ruido (para que no sea determinista)
        ctx = ContextoEnunciativo(temperatura_semantica=1.2)
        
        respuesta, _, info = colapso_parole(oraculo, ctx)
        
        print(f"🔮 El oráculo dice: {respuesta}")
        print(f"   (Confianza: {info['probabilidades_originales'][opciones.index(respuesta)]*100:.1f}%)")
        
        self.historial.append({
            "pregunta": pregunta,
            "respuesta": respuesta,
            "tiempo": time.time()
        })
        
        return respuesta
    
    def estadisticas(self):
        """Ver historial de respuestas"""
        if not self.historial:
            return "Sin consultas aún"
        
        print("\n=== HISTORIAL DEL ORÁCULO ===")
        for i, h in enumerate(self.historial[-5:], 1):
            print(f"{i}. {h['pregunta']} → {h['respuesta']}")
        return ""

# Crear el oráculo
oraculo = OraculoCuantico()

print("=== ORÁCULO SEMÁNTICO CUÁNTICO ===\n")
print("Un oráculo que responde colapsando superposiciones cuánticas\n")

oraculo.preguntar(
    "¿Qué me depara el futuro?",
    ["💰 riqueza", "💕 amor", "🏆 éxito", "🧘 paz", "🎭 sorpresa"]
)

oraculo.preguntar(
    "¿Cuál es el sentido de la vida?",
    ["❤️ amar", "🎨 crear", "📚 aprender", "🌱 crecer", "🎲 jugar"],
    sesgo=[1.5, 1.2, 1.0, 1.3, 1.8]  # Ligero sesgo hacia "jugar"
)

oraculo.preguntar(
    "¿Debo tomar el riesgo?",
    ["✅ sí", "❌ no", "🤔 tal vez", "🌀 depende"]
)

oraculo.estadisticas()

# 🔬 El Operador Diferencia

## El corazón matemático de la fusión Saussure-Cuántica

**Principio fundamental de Saussure:** *"En la lengua solo hay diferencias sin términos positivos"*

En este notebook exploramos el operador $\hat{D}$ que materializa computacionalmente esta idea: un signo ES por NO SER los demás signos del sistema.

## 1. Importar y configurar

````python
import sys
sys.path.insert(0, '..')

from saussure_quantum import SignoCuanto, Langue
from saussure_quantum.operators import (
    operador_diferencia, 
    OperadorDiferencia,
    principio_negatividad,
    similitud_diferencial
)
import numpy as np
import matplotlib.pyplot as plt

print("✅ Importado correctamente")
````

## 2. La intuición: ¿Qué significa "ser por no ser"?

Imagina tres fonemas: /p/, /b/, /t/

- **Aisladamente**: cada uno es una "cosa"
- **En el sistema de la lengua**: /p/ es lo que ES porque NO ES /b/ ni /t/

El operador diferencia transforma cada fonema aislado en un estado que representa **pura diferencia**.

````python
# Crear fonemas como estados puros (sin contexto)
fonema_p = SignoCuanto(["/p/", "/b/", "/t/"], [1, 0, 0])
fonema_b = SignoCuanto(["/p/", "/b/", "/t/"], [0, 1, 0])
fonema_t = SignoCuanto(["/p/", "/b/", "/t/"], [0, 0, 1])

print("=== FONEMAS AISLADOS (INGENUOS) ===")
print("Fonema /p/:", fonema_p)
print("Fonema /b/:", fonema_b)
print("Fonema /t/:", fonema_t)

print("\n=== APLICANDO OPERADOR DIFERENCIA ===")
diferencia = operador_diferencia([fonema_p, fonema_b, fonema_t])
print("Ser /p/ por NO SER /b/ ni /t/:")
print(diferencia)
````

## 3. Visualizando la diferencia

Observa cómo el operador diferencia distribuye la probabilidad entre todos los fonemas. Un fonema ya no es "una cosa", sino una **relación de oposiciones**.

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Antes del operador diferencia
probs_antes = [1.0, 0.0, 0.0]
axes[0].bar(["/p/", "/b/", "/t/"], probs_antes, color='lightblue')
axes[0].set_title('Antes: /p/ como "cosa"')
axes[0].set_ylim(0, 1)
axes[0].set_ylabel('Probabilidad')

# Después del operador diferencia
probs_despues = np.abs(diferencia.amplitudes)**2
axes[1].bar(["/p/", "/b/", "/t/"], probs_despues, color='coral')
axes[1].set_title('Después: /p/ como "diferencia"')
axes[1].set_ylim(0, 1)

plt.suptitle('El operador diferencia: de la sustancia a la relación', fontsize=12)
plt.tight_layout()
plt.show()

print("\n📊 Interpretación:")
print("  /p/ ahora tiene probabilidad {:.1f}% de colapsar a /p/,".format(probs_despues[0]*100))
print("  pero también {:.1f}% de colapsar a /b/ y {:.1f}% a /t/".format(probs_despues[1]*100, probs_despues[2]*100))
print("\n  → Esto es el 'ser por no ser': el fonema contiene a sus opuestos como posibilidad.")
````

## 4. Análisis de negatividad

La función `principio_negatividad()` cuantifica cuánto "debe" cada significante a su negación de los otros.

````python
analisis = principio_negatividad(diferencia)

print("=== ANÁLISIS DE NEGATIVIDAD ===\n")
print(f"Significante más probable: {analisis['significante_principal']}")
print(f"Negatividad total del sistema: {analisis['negatividad_total']:.3f}")
print("\nNegatividad por significante (cuánto 'debe' a no ser los otros):")

for sig, neg in analisis['negatividad_por_significante'].items():
    print(f"  {sig}: {neg:.3f}")

# Visualizar
fig, ax = plt.subplots(figsize=(8, 4))
significantes = list(analisis['negatividad_por_significante'].keys())
negatividades = list(analisis['negatividad_por_significante'].values())

ax.bar(significantes, negatividades, color='darkgreen', alpha=0.7)
ax.set_ylabel('Fuerza de negatividad')
ax.set_title('Cada signo ES por NO SER los otros')
ax.set_ylim(0, max(negatividades) * 1.1)

for i, v in enumerate(negatividades):
    ax.text(i, v + 0.02, f'{v:.3f}', ha='center')

plt.show()
````

## 9. Aplicación filosófica: Disolver la sustancia

**Antes de Saussure:** Creíamos que las palabras nombran cosas preexistentes.

**Después de Saussure:** Las palabras adquieren valor por oposición dentro de un sistema.

**Después de la fusión cuántica:** La "realidad" es un colapso desde un espacio de diferencias puras.

```python
# Una demostración conceptual
conceptos = ["verdad", "mentira", "duda", "certeza"]
print("=== DEMOSTRACIÓN: LA VERDAD ES DIFERENCIA ===\n")

# "Verdad" como estado aislado (ingenuo)
verdad_aislada = SignoCuanto(conceptos, [1, 0, 0, 0])

# El sistema completo de oposiciones
verdad = SignoCuanto(conceptos, [1, 0, 0, 0])
mentira = SignoCuanto(conceptos, [0, 1, 0, 0])
duda = SignoCuanto(conceptos, [0, 0, 1, 0])
certeza = SignoCuanto(conceptos, [0, 0, 0, 1])

sistema = operador_diferencia([verdad, mentira, duda, certeza])

print("Concepto 'verdad' como sustancia aislada:")
print(verdad_aislada)

print("\nConcepto 'verdad' en el sistema de diferencias (ser por no ser):")
print(sistema)

print("\n✨ Conclusión filosófica:")
print("  La verdad NO es una propiedad intrínseca de un enunciado.")
print("  La verdad EMERGE de su oposición a la mentira, la duda y la certeza.")
print("  Y colapsa en el acto de ser enunciada.")
